```python
import pandas as pd
import numpy as np
from IPython.display import Image
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.graphics.tsaplots import plot_acf
np.random.seed(1)
```

# Problem 1 

```python
# Creating a sample of 1000 datapoints from the ARMA(1,1) process which is MA(1) + AR(1)
n = 1000

#Values for Phi and Theta from the book's page 484
phi = [0.6,0.6,-0.8,0.5]
theta = [0.4,-0.4,0.55,-0.9]

for i in range(len(phi)):
    y = [1] 
    errors = [np.random.normal(0,1)] #distribution of the errors is assumed to be Standard Normal
    for _ in range(1,n):
        errors.append(np.random.normal(0,1))
        y.append( phi[i]*y[-1] + errors[-1] + theta[i]*errors[-2] )
    y = np.array(y) #cast to an array from List datatype as above 
    
    plot_acf(y) #apply the autocorrelation plotting function for stats model
    plt.title('Phi={},Theta={}'.format(phi[i],theta[i]))
```
